```python
import numpy as np
import math


array = np.arange(0, 128, 1).reshape((2, 2, 2, 2, 2, 2, 2)) #создание тензора размерности 7
array -= np.arange(127, -1, -1).reshape((2, 2, 2, 2, 2, 2, 2)) #математическая операция над тензором 
print(array.shape) #размерность тензора
print(array.dtype) #тип данных тензора
array #вывод результата
```

```python
E = np.eye(3) #создание единичной матрицы
M = np.array([[8, 32, 47], [7, 22, 81], [63, 91, 5]]) #создание матрицы
print("Сложение матрицы со скаляром: ", M + 5*E, sep='\n') #Сложение матрицы со скаляром по правилам линейной алгебры
print("Почленное сложение элементов матрицы со скаляром: ", M + 5, sep='\n') #Почленное сложение элементов матрицы со скаляром
```

```python
M = np.array([[8, 32, 47], [7, 22, 81], [63, 91, 5]])
print("Матрица:", M, sep="\n")
print("Определитель матрицы:", np.linalg.det(M)) #нахождение оприделителя матрицы
print("Проверка методом разложения миноров:")
M11, M12, M13 = M[0, 0], M[0, 1], M[0, 2] #получение элементов с помощью индексирования
print(f'M11, M12, M13: {M[0, 0]}, {M[0, 1]}, {M[0, 2]}')
#нахождение миноров с помощью индексирования
det11, det12, det13 = math.floor(np.linalg.det(M[1:,1:])), math.floor(np.linalg.det(M[1:, [0, 2]])), math.floor(np.linalg.det(M[1:, :2]))
print(f'det11, det12, det13: {np.linalg.det(M[1:,1:])}, {np.linalg.det(M[1:, [0, 2]])}, {np.linalg.det(M[1:, :2])}')
print(f'M11*det11 - M12*det12 + M13*det13 = {M11*det11 - M12*det12 + M13*det13}') #нахождение определителя с помощью разложения по минорам
print('68885 = 68885 => всё чётенько') #вывод
```
